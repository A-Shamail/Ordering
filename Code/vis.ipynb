```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, FancyArrowPatch
from matplotlib.patches import Arc
```

```python
# Set up plot
fig, ax = plt.subplots(figsize=(7, 7))

# Create horizontal oval (red cloud)
horizontal_oval = Ellipse(
    xy=(0.02, 0.0),   # center at (0.02, 0)
    width=0.04,       # wider horizontally
    height=0.006,     # very skinny vertically
    angle=0,
    color='red',
    alpha=0.2,
    label='Horizontal cloud'
)
ax.add_patch(horizontal_oval)

# Create vertical oval (blue cloud)
vertical_oval = Ellipse(
    xy=(0.0, 0.02),   # center at (0, 0.02)
    width=0.006,      # very skinny horizontally
    height=0.04,      # tall vertically
    angle=0,
    color='blue',
    alpha=0.2,
    label='Vertical cloud'
)
ax.add_patch(vertical_oval)

# Generate random points inside each cloud
np.random.seed(42)

# Points for red cloud (around (0.02, 0))
x_red = np.random.normal(loc=0.03, scale=0.005, size=30)
y_red = np.random.normal(loc=0.0, scale=0.0008, size=30)

# Points for blue cloud (around (0, 0.02))
x_blue = np.random.normal(loc=0.0, scale=0.0008, size=30)
y_blue = np.random.normal(loc=0.03, scale=0.004, size=30)

angle_center = (0, 0)   # center at origin
theta1 = 0              # starting angle (X-axis)
theta2 = 90             # ending angle (example: 30 degrees for demo)
arc_radius = 0.003       # size of the arc

# Add the arc showing the angle
arc = Arc(angle_center, width=2*arc_radius, height=2*arc_radius,
          angle=0, theta1=theta1, theta2=theta2, color='black', lw=1.5)
ax.add_patch(arc)

# Label theta
ax.text(0.0022, 0.003, r'$\theta$', fontsize=14, color='black')

# Plot points
ax.scatter(x_red, y_red, color='red', s=30)
ax.scatter(x_blue, y_blue, color='blue', s=30)

ax.annotate("A adds info,\nB adds little after", xy=(0.03, 0.002), xytext=(0.03, 0.016),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=16, color='black')

ax.annotate("B adds info,\nA adds little after", xy=(0.002, 0.03), xytext=(0.005, 0.04),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=16, color='black')


ax.set_xticks([])  # removes x-axis values
ax.set_yticks([])  # removes y-axis values



# Axes, Labels
ax.axhline(0, color='gray', linestyle='--', linewidth=1)
ax.axvline(0, color='gray', linestyle='--', linewidth=1)
ax.set_xlim(-0.01, 0.05)
ax.set_ylim(-0.01, 0.05)
ax.set_xlabel("MSE Reduction after adding A", fontsize=16)
ax.set_ylabel("MSE Reduction after adding B", fontsize=16)
ax.set_title("Redundancy between A and B (A  B)", fontsize=17)
# remove grid in background
# ax.legend()
ax.grid(False)
plt.tight_layout()
plt.savefig('fig1_redundancy.pdf',
            bbox_inches='tight',  # removes extra white spaces around the figure
            dpi=300,             # sets resolution
            format='pdf')  


plt.show()
```
