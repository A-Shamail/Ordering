```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, FancyArrowPatch
from matplotlib.patches import Arc
```

```python
# Set up plot
fig, ax = plt.subplots(figsize=(7, 7))

# Create horizontal oval (red cloud)
horizontal_oval = Ellipse(
    xy=(0.02, 0.0),   # center at (0.02, 0)
    width=0.04,       # wider horizontally
    height=0.006,     # very skinny vertically
    angle=0,
    color='red',
    alpha=0.3,
    label='Horizontal cloud'
)
ax.add_patch(horizontal_oval)

# Create vertical oval (blue cloud)
vertical_oval = Ellipse(
    xy=(0.0, 0.02),   # center at (0, 0.02)
    width=0.006,      # very skinny horizontally
    height=0.04,      # tall vertically
    angle=0,
    color='blue',
    alpha=0.3,
    label='Vertical cloud'
)
ax.add_patch(vertical_oval)

# Generate random points inside each cloud
np.random.seed(42)

# Points for red cloud (around (0.02, 0))
x_red = np.random.normal(loc=0.03, scale=0.005, size=30)
y_red = np.random.normal(loc=0.0, scale=0.0008, size=30)

# Points for blue cloud (around (0, 0.02))
x_blue = np.random.normal(loc=0.0, scale=0.0008, size=30)
y_blue = np.random.normal(loc=0.03, scale=0.005, size=30)

angle_center = (0, 0)   # center at origin
theta1 = 0              # starting angle (X-axis)
theta2 = 90             # ending angle (example: 30 degrees for demo)
arc_radius = 0.003       # size of the arc

# Add the arc showing the angle
arc = Arc(angle_center, width=2*arc_radius, height=2*arc_radius,
          angle=0, theta1=theta1, theta2=theta2, color='black', lw=1.5)
ax.add_patch(arc)

# Label theta
ax.text(0.0022, 0.003, r'$\theta$', fontsize=12, color='black')

# Plot points
ax.scatter(x_red, y_red, color='red', s=30)
ax.scatter(x_blue, y_blue, color='blue', s=30)

ax.annotate("X1 then X2 added", xy=(x_red[7], y_red[7]), xytext=(x_red[0]+0.005, y_red[0]+0.01),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=9, color='black')

ax.annotate("X2 then X1 added", xy=(x_blue[11], y_blue[11]), xytext=(x_blue[0]+0.005, y_blue[0]+0.01),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=9, color='black')


ax.text(
    0.0339, 0.04924,  # (x, y) coordinates in figure (adjust if needed)
    r"Red Skinnyness = 0.98" + "\n" +
    r"Red Angle = $-3.1$ rad" + "\n" +
    r"Blue Skinnyness = 0.95" + "\n" +
    r"Blue Angle = $-1.6$ rad" + "\n" +
    r"L-Score = 0.929",
    fontsize=9,
    verticalalignment='top',
    bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5', alpha=0.0)
)





# Axes, Labels
ax.axhline(0, color='gray', linestyle='--', linewidth=1)
ax.axvline(0, color='gray', linestyle='--', linewidth=1)
ax.set_xlim(-0.01, 0.05)
ax.set_ylim(-0.01, 0.05)
ax.set_xlabel("MSE Reduction after adding X1", fontsize=12)
ax.set_ylabel("MSE Reduction after adding X2", fontsize=12)
ax.set_title("L-Pattern with Clouds and Points")
# remove grid in background
# ax.legend()
ax.grid(False)
plt.tight_layout()
plt.savefig('fig1.pdf',
            bbox_inches='tight',  # removes extra white spaces around the figure
            dpi=300,             # sets resolution
            format='pdf')  


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up figure
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Define different skinnyness levels
skinnyness_settings = [
    (1.0, 1.0, "Skinnyness  0.5"),    # round
    (3.0, 1.0, "Skinnyness  0.75"),   # medium stretched
    (10.0, 1.0, "Skinnyness  0.9")    # very skinny
]

np.random.seed(42)

for ax, (stretch_x, stretch_y, label) in zip(axes, skinnyness_settings):
    # Generate points
    x = np.random.normal(0, stretch_x, size=300)
    y = np.random.normal(0, stretch_y, size=300)
    
    # Plot
    ax.scatter(x, y, s=10, alpha=0.7)
    ax.set_title(label, fontsize=12)
    ax.axhline(0, color='gray', linestyle='--', linewidth=1)
    ax.axvline(0, color='gray', linestyle='--', linewidth=1)
    ax.set_xlim(-15, 15)
    ax.set_ylim(-15, 15)
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])

fig.suptitle("Effect of Skinnyness on Point Clouds", fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.85)

# Save the figure
plt.savefig('skinnyness_examples.pdf', bbox_inches='tight', dpi=300, format='pdf')
plt.show()
```
