```python
def generate_dataset(n_samples=2000):
    np.random.seed(42)

    # Generate independent features
    X1 = np.random.uniform(-1, 1, n_samples)

    Y = X1

    # Create redundancies
    X2 = np.log(np.abs(np.sin(X1)) + 1e-5)  # Double transformation
    X3 = np.abs(X1)  # Piecewise
    X4 = np.maximum(0, X1)  # ReLU
    X5 = np.sin(X1 + np.pi / 4)  # Periodic
    X6 = 1 / (X1 + 1e-5)  # Reciprocal transformation

    df = pd.DataFrame({'X1': X1, 'X2': X2, 'X3': X3, 'X4': X4, 'X5': X5, 'X6': X6, 'Y': Y})
    
    return df

# Example dataset with exponential dependency
# df_exp = generate_dataset(dependency_type="exponential")
df = generate_dataset()
display(df.head())
```
